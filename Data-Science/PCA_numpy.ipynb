```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randint(0, 100, 5000).reshape(1000, 5)
```

```python
X.shape
```

```python
covariance_mat = np.cov(X, rowvar=False)
```

```python
covariance_mat, covariance_mat.shape
```

```python
eigen_value, eigen_vector = np.linalg.eigh(covariance_mat)
```

```python
print(eigen_value)
print(eigen_vector, eigen_vector.shape)
```

```python
sort_index = np.argsort(eigen_value)
```

```python
sort_index = sorted(sort_index, reverse=True)
```

```python
eigen_value = eigen_value[sort_index]
eigen_vector = eigen_vector[sort_index, :]
```

```python
eigen_value, eigen_vector
```
